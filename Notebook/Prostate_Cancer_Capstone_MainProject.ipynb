```python
df = df.drop(['AGE_ADJUSTED_CI_LOWER','AGE_ADJUSTED_CI_UPPER','AGE_ADJUSTED_RATE', 
              'CI_LOWER','CI_UPPER','CRUDE_CI_LOWER','CRUDE_CI_UPPER','CRUDE_RATE','ICCC_GROUP'], axis = 1) #Dropping not useful Variables
```

```python
#Getting Unique Value count from each Column
df.nunique()[:10]
```

```python
#Replacing ~ Value 
df= df.replace(["~",".","+","-"],np.nan)
len(df)
```

```python
#Counting NULL Values in Each column
df.isna().sum()
```

```python
#Converting to category
df['AGE'] = df['AGE'].astype('category')
df["Age_cat"] = df["AGE"].cat.codes
df['SEX'] = df['SEX'].astype('category')
df["SEX_cat"] = df["SEX"].cat.codes
print(df['SEX_cat'].value_counts()) #-1 for NaN
```

```python
#Getting top 20 types plotted cancer types/sites

df = df.replace(["~",".","+","-",None],0)
df['COUNT'] = df['COUNT'].astype(int)
Site = df.groupby(['SITE'], as_index=False)['COUNT'].sum()
Site=Site.sort_values(by=['COUNT'], ascending= False)
print(Site.head(20))
g = sns.barplot(x = 'SITE', y = 'COUNT', data = Site[2:30])
plt.xticks(rotation = 90);
#Eliminating First two rows while plotting as they include all types
```

```python
#Getting top 20 types plotted cancer sites for FEMALE

Site = df.groupby(['SEX','SITE'], as_index=False)['COUNT'].sum()
Site = Site[Site['SEX'] == 'Female']
Site=Site.sort_values(by=['COUNT'], ascending= False)
print(Site.head(20))

g = sns.barplot(x = 'SITE', y = 'COUNT', data = Site[2:30])
plt.xticks(rotation = 90);

#Eliminating First two rows while plotting as they include all types
```

```python
#Getting top 30 types plotted cancer sites for MALE

Site = df.groupby(['SEX','SITE'], as_index=False)['COUNT'].sum()
Site = Site[Site['SEX'] == 'Male']
Site=Site.sort_values(by=['COUNT'], ascending= False)
print(Site.head(20))

g = sns.barplot(x = 'SITE', y = 'COUNT', data = Site[2:32])
plt.xticks(rotation = 90);

#Eliminating First two rows while plotting as they include all types
```
