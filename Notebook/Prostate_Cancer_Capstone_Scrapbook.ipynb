```python
import pandas as pd
import seaborn as sns
```

```python

countyData=pd.read_csv('BYAREA_COUNTY.TXT',delimiter="|")
```

```python
countyData.head()
```

```python
sns.set(rc={'figure.figsize':(17.7,8.27)})
sns.countplot(x="STATE",data=countyData, color ='Blue')
```

```python
sns.countplot(x="SEX",data=countyData)
```

```python
sns.countplot(x="STATE",data = countyData, hue="SEX")
```

```python
sns.countplot(x="STATE",data = countyData, hue="EVENT_TYPE")
```

```python
sns.countplot(x="RACE",data=countyData)
```

```python
ax = sns.countplot(x="SITE",data=countyData)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
```

```python
prostateCancer = countyData.loc[countyData['SITE']=="Prostate"]
prostateCancer.head()
```

```python
sns.countplot(x="SEX",data=prostateCancer)
```

```python
sns.countplot(x= "RACE", data = prostateCancer)
```

```python
sns.countplot(x="STATE",data=prostateCancer)
```

```python
sns.countplot(x="YEAR",data = prostateCancer)
```

```python
sns.countplot(x="YEAR", hue="EVENT_TYPE",data=prostateCancer)
```
